```python
wind = []
for i in soupq9.find_all('td',class_='wind'):
    wind.append(i.text)

wind1 = [wind[x] for x in range(6,30)]
```

```python
q10 = requests.get('https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/')
q10
```

```python
soupq10 = BeautifulSoup(q10.content, 'html.parser')
```

```python
data = soupq10.find_all('div',class_="blog--single__content column--3-4 u-spacing-third")

temp=[]

for i in data:

    for j in i.find_all('p'):

        t = j.text.replace('\n', '') # only extracting data

        temp.append(t)
        
name=[]

for i in range(1,30,3):

    name.append(temp[i])

name
```

```python
desc = []
for i in range(2,30,3):

    desc.append(temp[i])
    
desc
```

```python
imagesURL = soupq10.find_all('img', {'src':re.compile('.jpg')})
imagesURL.pop(0)

for i in imagesURL:
    print(i['src']+'\n')
```

```python
print(len(name),len(desc),len(imagesURL))
```
